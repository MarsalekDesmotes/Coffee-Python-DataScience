**14.** 40 İLE 49 YAŞ ARASINDAKİ İNSANLARIN NE SIKLIKLA KAHVE İÇTİĞİNİN LİSTESİNİNİ GÖSTEREN SORGU:




```python
data.loc[(data['AgeRange']=="40 to 49") & (data['CoffeeCupsPerDay'])]
```

```python
data.loc[(data['CodingHours']<3) & (data['Gender'] =="Male")]
```

**VERİ GÖRSELLEŞTİRMESİ**

Veri Görselleştirmesi yapabilmek için matplatlib kütüphanesini ve modüllerini import ediyoruz.

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

Mathplotlibde değerleri string olarak algılamaması için float'a çeviriyoruz.

```python
data['CodingHours'] = data['CodingHours'].astype('float')
data['CoffeeCupsPerDay'] = data['CoffeeCupsPerDay'].astype('float')

```

Mathplotlib kütüphanesinde veri görselleştirmek için kullanılan bazı komutlar:



*   **plt.plot()**: Grafiğin ana komutunu oluşturur. Bu komutun içine ilk önce x ekseni, sonrasında y ekseni yazılır. Grafikteki temel bazı değişkenleri değiştirmek istersek ilgili kodu buraya yazmamız gerekir. Örneğin grafiğin rengi, çizgi kalınlığı, aralıklar vs.
*   **figsize:**Grafiğin boyutunu belirler. Yaygın olarakplt.figure(figsize=(8,6)) şeklinde kullanılır. Farklı kullanımları da mevcuttur. Parantez içindeki ilk yazılan değer grafiğin eni, ikincisi boyudur.
*  **plt.xlabel():**  x ekseninin başlığını oluşturur.
*   **plt.ylabel()**: y ekseninin başlığını oluşturur.
*  **plt.title():** Grafiğin başlığını oluşturur.

*  **plt.show():** Grafiği çizdirmemizi sağlar. Bu kodu yazmazsak grafik çizilir fakat bazen ekranda gözükmeyebilir.


Bu komutları teker teker grafik üzerinde gösterelim.





```python
plt.figure(figsize=(12,6))

plt.plot(data.Gender,data.CoffeeSolveBugs) #biz belirlemediğimiz sürece rengi otomatik kendisi verir. 

plt.show()
```
